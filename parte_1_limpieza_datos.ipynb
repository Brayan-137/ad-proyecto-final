## Contexto del juego

Los partidos de pádel registrados en este dataset corresponden a una variante recreativa
jugada por estudiantes universitarios. Las reglas de puntuación son las siguientes:

- El partido se juega en un formato de puntos directos (sin games ni sets).
- Gana el primer equipo en alcanzar **7 puntos** con una diferencia mínima de **2 puntos**.
- Si al llegar a 7 puntos no hay diferencia de 2, el partido continúa hasta
  que algún equipo la consiga.
- Cada punto es disputado por dos equipos de dos jugadores cada uno.
- Las posiciones de los jugadores en la cancha (LEFT/RIGHT, TOP/BOTTOM)
  pueden variar entre puntos dentro del mismo partido.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import unicodedata
warnings.filterwarnings("ignore")

# Cargar el dataset
data_videos = pd.read_csv("./databases/2026-1/Base_Videos_Final_202601.csv")
data_partidos = pd.concat([
    pd.read_excel("./databases/2026-1/Traza_Datos_Partidos_202601.xlsx", sheet_name="CANCHA_1"),
    pd.read_excel("./databases/2026-1/Traza_Datos_Partidos_202601.xlsx", sheet_name="CANCHA_2")
], ignore_index=True)
data_jugadores = pd.read_excel("./databases/2026-1/Data_Jugadores_202601.xlsx")
```

```python
def normalizar_nombre(nombre):
    if pd.isna(nombre):
        return nombre
    # Minúsculas, strip, title
    nombre = str(nombre).strip().lower().title()
    # Quitar tildes
    nombre = unicodedata.normalize('NFKD', nombre).encode('ascii', 'ignore').decode('utf-8')
    return nombre

# Normalizar en data_videos
data_videos['player_name'] = data_videos['player_name'].apply(normalizar_nombre)

# Normalizar en data_partidos
cols_jugadores = [
    'ID_JUGADOR_1_EQUIPO_1', 'ID_JUGADOR_2_EQUIPO_1',
    'ID_JUGADOR_1_EQUIPO_2', 'ID_JUGADOR_2_EQUIPO_2'
]
for col in cols_jugadores:
    data_partidos[col] = data_partidos[col].apply(normalizar_nombre)

# Normalizar en data_jugadores
data_jugadores['NOMBRE'] = data_jugadores['NOMBRE'].apply(normalizar_nombre)
```

**Hallazgos**

- Hay frames en los cuales el equipo no es detectado team (la posición del equipo top or bottom) -> Al no cambiar durante la duración del punto el partido, se reemplazara con la media de ese punto o partido
- Hay casos en los que la distancia al compañero del jugador esta calculada en solo uno de los dos miembros del equipo -> Revisar la existencia de esos casos y corregirlo; En caso de que no pero se tengan las posiciones de los dos jugadores calcularlo para amos
- Los valores nulos de las variables prev_x, prev_y, player_displacement, player_speed, prev_speed, player_acceleration, ball_position_x_prev, ball_position_y_prev, ball_displacement, ball_speed_mps, son nulos en los primeros frames de cada punto -> Se plantea eliminar las filas de estos primeros frames
- Los valores nulos de las variables ball_position_x_prev, ball_position_y_prev, ball_displacement, ball_speed_mps que no son el primer frame no son calculadas -> Revisar si se cuenta con las posiciones de la pelota y corregirlo
- time_since_last_hit es nulo si al principio de cada punto hasta que no se golpee por primera vez -> imputar con -1, indica que no ha sido golpeado todavia
- tras una primera limpieza de los primeros frames quedarón apenas 285 nulos en las variables de ball_position_x_prev, ball_position_y_prev, ball_displacement, ball_speed_mps -> al ser tan pocos se opto por su eliminación
- en la player_name aparecen nombres Player_X (con X siendo diferentes números), esta es la forma del programa de power vision de registrar los jugadores que no pudo reconocer -> imputar este nombre y el id del jugador con el id del punto, el partido, y por el frame id, ya que por cada punto y partido debería haber 4 frame con el mismo id, entonces por descarte se puede saber de que jugar es ese id, en caso de que no se pueda hacer por descarte (que en un frame no se detecten 4 jugadors) implica que en ese frame hay muy poca información y al no ser un número significativo de ellos (9716) se pueden eliminar
- para la variable zone, aparecen 3 valores left_box, right_box y back_court, pero los videos se tomaron siempre cuando los jugadores estaban dentro de los límites de la cancha por lo cual esta es la forma del programa de decir que no los esta detectando -> imputar por partido y id del jugador
- después de una primera iteración para imputar los datos de zone, hubieron 1275 registros nulos, en los cuales no hubo una moda con la cual imputar los datos -> al ser muy pocos frames del computo global, aunque se podrían haber eliminados, se decidio calcular la zona a la que suelen entrar, que aunque este errada no tendrá un gran impacto y se podrá conservar el resto de la información

**Tratamiento**

```python
# ============================================================
# TRATAMIENTO DE VALORES NULOS - data_videos (VERSIÓN CORREGIDA)
# ============================================================

# ------------------------------------------------------------
# 1. Eliminar primeros frames de cada punto
#    (prev_x, prev_y, player_displacement, player_speed,
#     prev_speed, player_acceleration y variables de pelota
#     son nulos estructuralmente en el primer frame)
# ------------------------------------------------------------
data_videos = data_videos[data_videos['frame_idx'] != 0].reset_index(drop=True)

data_videos = data_videos.dropna(
    subset=['prev_speed', 'player_acceleration_mps2']
).reset_index(drop=True)

# ------------------------------------------------------------
# 2. Imputar time_since_last_hit con -1
#    (nulo hasta que ocurre el primer golpe del punto)
# ------------------------------------------------------------
data_videos['time_since_last_hit'] = data_videos['time_since_last_hit'].fillna(-1)

# ------------------------------------------------------------
# 3. Imputar team con moda por partido y jugador
#    (no cambia durante el punto/partido)
# ------------------------------------------------------------
def imputar_moda(series):
    moda = series.dropna().mode()
    return series.fillna(moda[0] if not moda.empty else np.nan)

# Nivel 1: moda por partido y jugador
data_videos['team'] = (
    data_videos
    .groupby(['partido', 'player_id'])['team']
    .transform(imputar_moda)
)

# Nivel 2: moda global por jugador
data_videos['team'] = (
    data_videos
    .groupby('player_name')['team']
    .transform(imputar_moda)
)

data_videos = data_videos.dropna(subset=['team']).reset_index(drop=True)

# ------------------------------------------------------------
# 4. Imputar player_name y player_id de Player_X por descarte
#    Lógica: en cada (partido, punto, frame_idx) deben aparecer
#    4 jugadores. Si hay exactamente 1 Player_X y 3 conocidos,
#    se puede identificar al desconocido por descarte.
# ------------------------------------------------------------
jugadores_por_partido = (
    data_partidos[[
        'ID_PARTIDO',
        'ID_JUGADOR_1_EQUIPO_1', 'ID_JUGADOR_2_EQUIPO_1',
        'ID_JUGADOR_1_EQUIPO_2', 'ID_JUGADOR_2_EQUIPO_2'
    ]]
    .drop_duplicates('ID_PARTIDO')
    .melt(id_vars='ID_PARTIDO', value_name='player_name')
    .drop(columns='variable')
    .rename(columns={'ID_PARTIDO': 'partido'})
)
jugadores_por_partido['player_name'] = jugadores_por_partido['player_name'].apply(normalizar_nombre)

corregidos_playerx = 0
excepciones_playerx = 0

mask_playerx = data_videos['player_name'].str.contains(r'^Player_\d+$', na=False)
frames_playerx = data_videos[mask_playerx][['partido', 'punto', 'frame_idx']].drop_duplicates()

for _, row in frames_playerx.iterrows():
    partido, punto, frame = row['partido'], row['punto'], row['frame_idx']

    mask_frame = (
        (data_videos['partido'] == partido) &
        (data_videos['punto'] == punto) &
        (data_videos['frame_idx'] == frame)
    )

    jugadores_frame = data_videos[mask_frame]
    playerx_en_frame = jugadores_frame[
        jugadores_frame['player_name'].str.contains(r'^Player_\d+$', na=False)
    ]

    # Solo corregir si hay exactamente 1 Player_X en el frame
    if len(playerx_en_frame) != 1:
        excepciones_playerx += len(playerx_en_frame)
        continue

    conocidos = jugadores_frame[
        ~jugadores_frame['player_name'].str.contains(r'^Player_\d+$', na=False)
    ]['player_name'].tolist()

    jugadores_partido = jugadores_por_partido[
        jugadores_por_partido['partido'] == partido
    ]['player_name'].tolist()

    faltante = [j for j in jugadores_partido if j not in conocidos]

    if len(faltante) == 1:
        idx = playerx_en_frame.index[0]
        data_videos.loc[idx, 'player_name'] = faltante[0]
        corregidos_playerx += 1
    else:
        excepciones_playerx += 1

print(f"Player_X corregidos por descarte:       {corregidos_playerx}")
print(f"Player_X sin corrección (excepciones):  {excepciones_playerx}")

# Diagnóstico de excepciones por partido
total_frames_playerx = mask_playerx.sum()
print(f"\nTotal frames con Player_X:              {total_frames_playerx:,}")
# CORRECCIÓN: el porcentaje sobre Player_X ahora especifica que son los NO corregidos
print(f"% de Player_X no corregidos:            {excepciones_playerx / total_frames_playerx * 100:.1f}%")
print(f"% de Player_X corregidos:               {corregidos_playerx / total_frames_playerx * 100:.1f}%")

excepciones_por_partido = (
    data_videos[
        mask_playerx &
        data_videos['player_name'].str.contains(r'^Player_\d+$', na=False)
    ]
    .groupby('partido')
    .size()
    .reset_index(name='frames_excepcion')
    .sort_values('frames_excepcion', ascending=False)
)
total_por_partido = (
    data_videos
    .groupby('partido')
    .size()
    .reset_index(name='total_frames')
)
excepciones_por_partido = excepciones_por_partido.merge(total_por_partido, on='partido')
excepciones_por_partido['pct'] = (
    excepciones_por_partido['frames_excepcion'] /
    excepciones_por_partido['total_frames'] * 100
).round(2)
print(f"\nDistribución de excepciones por partido:")
print(excepciones_por_partido.to_string(index=False))

# Eliminar frames de Player_X no corregibles
# CORRECCIÓN: medir filas antes/después en lugar de usar mask original
filas_antes_playerx = len(data_videos)
mask_playerx_residual = data_videos['player_name'].str.contains(r'^Player_\d+$', na=False)
data_videos = data_videos[~mask_playerx_residual].reset_index(drop=True)

print(f"\nFrames eliminados: {filas_antes_playerx - len(data_videos):,}")
print(f"Shape final:       {data_videos.shape}")

# ------------------------------------------------------------
# 5. Imputar zone
#    - left_box / right_box → left / right
#    - back_court y nulos → moda de left/right por partido y jugador
#      Si no hay moda por partido, usar moda global por jugador
# ------------------------------------------------------------

# Normalizar left_box/right_box → left/right
data_videos['zone'] = data_videos['zone'].replace({
    'left_box':  'left',
    'right_box': 'right'
})

# Marcar back_court como NaN para imputar junto con los nulos reales
data_videos['zone'] = data_videos['zone'].replace({'back_court': np.nan})

def moda_zone(series):
    moda = series.dropna().mode()
    return moda[0] if not moda.empty else np.nan

# Nivel 1: moda por partido y jugador
# CORRECCIÓN: drop_duplicates para evitar duplicados en el merge
moda_por_jugador = (
    data_videos
    .groupby(['partido', 'player_id'])['zone']
    .apply(moda_zone)
    .reset_index()
    .rename(columns={'zone': 'zone_moda'})
    .drop_duplicates(subset=['partido', 'player_id'])
)

data_videos = data_videos.merge(moda_por_jugador, on=['partido', 'player_id'], how='left')

mask_nulo_n1 = data_videos['zone'].isna()
data_videos.loc[mask_nulo_n1, 'zone'] = data_videos.loc[mask_nulo_n1, 'zone_moda']
data_videos = data_videos.drop(columns='zone_moda')

print(f"Zone — valores imputados nivel 1 (partido+jugador): {mask_nulo_n1.sum()}")

# Nivel 2: moda global por jugador (para los que nunca tuvieron zona en ese partido)
moda_global = (
    data_videos
    .groupby('player_name')['zone']
    .apply(moda_zone)
    .reset_index()
    .rename(columns={'zone': 'zone_moda_global'})
)

data_videos = data_videos.merge(moda_global, on='player_name', how='left')

# CORRECCIÓN: recalcular mask_nulo después del nivel 1
mask_nulo_n2 = data_videos['zone'].isna()
data_videos.loc[mask_nulo_n2, 'zone'] = data_videos.loc[mask_nulo_n2, 'zone_moda_global']
data_videos = data_videos.drop(columns='zone_moda_global')

print(f"Zone — valores imputados nivel 2 (global por jugador): {mask_nulo_n2.sum()}")

# Nivel 3: eliminar si aún quedan nulos (jugador sin zona en ningún partido)
nulos_zone_residuales = data_videos['zone'].isna().sum()
if nulos_zone_residuales > 0:
    print(f"Eliminando {nulos_zone_residuales} filas con zone irrecuperable")
    data_videos = data_videos.dropna(subset=['zone']).reset_index(drop=True)

print(f"Distribución zone:\n{data_videos['zone'].value_counts()}")
print(f"Nulos restantes en zone: {data_videos['zone'].isna().sum()}")

# ------------------------------------------------------------
# 6. Imputar distance_player_to_teammate_m
#    Si un miembro del equipo tiene la distancia y el otro no,
#    copiar el valor. Si ninguno la tiene pero se tienen las
#    posiciones, calcularla para ambos.
# ------------------------------------------------------------
data_videos = data_videos.sort_values(
    ['partido', 'punto', 'frame_idx', 'team', 'player_id']
).reset_index(drop=True)

grp_equipo = data_videos.groupby(['partido', 'punto', 'frame_idx', 'team'])

companero_x = grp_equipo['player_position_x'].transform(
    lambda g: g.iloc[::-1].values
)
companero_y = grp_equipo['player_position_y'].transform(
    lambda g: g.iloc[::-1].values
)
dist_companero_espejo = grp_equipo['distance_player_to_teammate_m'].transform(
    lambda g: g.iloc[::-1].values
)

distancia_calculada = np.sqrt(
    (data_videos['player_position_x'] - companero_x)**2 +
    (data_videos['player_position_y'] - companero_y)**2
)

# Primero: copiar del compañero si él la tiene
data_videos['distance_player_to_teammate_m'] = (
    data_videos['distance_player_to_teammate_m']
    .fillna(dist_companero_espejo)
)

# Segundo: calcular con posiciones si aún es nulo
data_videos['distance_player_to_teammate_m'] = (
    data_videos['distance_player_to_teammate_m']
    .fillna(distancia_calculada)
)

# CORRECCIÓN: eliminar filas donde no se pudo calcular la distancia al compañero
# (jugador sin compañero detectable en ese frame)
nulos_distancia = data_videos['distance_player_to_teammate_m'].isna().sum()
if nulos_distancia > 0:
    print(f"Eliminando {nulos_distancia} filas sin compañero detectable")
    data_videos = data_videos.dropna(
        subset=['distance_player_to_teammate_m']
    ).reset_index(drop=True)

print(f"\nShape después del paso 6: {data_videos.shape}")

# ------------------------------------------------------------
# 7. Recalcular variables de pelota donde sean nulas
#    (frames que no son primeros pero tienen nulos)
# ------------------------------------------------------------
data_videos = data_videos.sort_values(
    ['partido', 'punto', 'player_id', 'frame_idx']
).reset_index(drop=True)

grp = data_videos.groupby(['partido', 'punto', 'player_id'])

# CORRECCIÓN: usar ffill tras shift(1) para cubrir nulos en frames consecutivos
data_videos['ball_position_x_prev'] = data_videos['ball_position_x_prev'].fillna(
    grp['ball_position_x'].transform(lambda x: x.shift(1).ffill())
)
data_videos['ball_position_y_prev'] = data_videos['ball_position_y_prev'].fillna(
    grp['ball_position_y'].transform(lambda x: x.shift(1).ffill())
)

ball_disp_calculado = np.sqrt(
    (data_videos['ball_position_x'] - data_videos['ball_position_x_prev'])**2 +
    (data_videos['ball_position_y'] - data_videos['ball_position_y_prev'])**2
)
data_videos['ball_displacement'] = data_videos['ball_displacement'].fillna(
    ball_disp_calculado
)
data_videos['ball_speed_mps'] = data_videos['ball_speed_mps'].fillna(
    data_videos['ball_displacement'] / data_videos['duration']
)

# CORRECCIÓN: medir filas eliminadas correctamente (filas, no celdas)
filas_antes_pelota = len(data_videos)
data_videos = data_videos.dropna(subset=[
    'ball_position_x_prev', 'ball_position_y_prev',
    'ball_displacement', 'ball_speed_mps'
]).reset_index(drop=True)

print(f"\nFilas eliminadas por nulos residuales de pelota: {filas_antes_pelota - len(data_videos)}")

# ------------------------------------------------------------
# Verificación final
# ------------------------------------------------------------
nulos = data_videos.isnull().sum()
nulos_restantes = nulos[nulos > 0]

if nulos_restantes.empty:
    print("\nSin nulos restantes")
else:
    print("\nNulos restantes por columna:")
    print(nulos_restantes)

print(f"\nShape final: {data_videos.shape}")
```

##### Outliers

**Detección**

Para el detección de los outliers se realizó una comparación utilizando el rango inerquartilico y unos límitos físicos previamente definidos, esto debido a que varias variables como player_speed_mps, player_acceleration_mps2, ball_speed_mps, ball_displacement y player_displacement en su mayoría de valores son iguales a cero afectando su capacidad para definir unos límites inferiores y superiore fiables, como se puede ver a continuación:

```python
# ============================================================
# DETECCIÓN DE OUTLIERS - Comparación IQR vs Límites Físicos
# ============================================================

vars_numericas = [
    'player_speed_mps', 'player_acceleration_mps2',
    'ball_speed_mps', 'ball_displacement',
    'distance_player_to_ball_m', 'distance_player_to_net_m',
    'distance_player_to_teammate_m', 'player_displacement'
]

limites_fisicos = {
    'player_speed_mps':              (0, 7),
    'player_acceleration_mps2':      (-10, 10),
    'ball_speed_mps':                (0, 50),
    'ball_displacement':             (0, 20),
    'distance_player_to_ball_m':     (0, 20),
    'distance_player_to_net_m':      (0, 10),
    'distance_player_to_teammate_m': (0, 10),
    'player_displacement':           (0, 2),
}

resultados = []

for col in vars_numericas:
    # --- IQR (3x) ---
    Q1 = data_videos[col].quantile(0.25)
    Q3 = data_videos[col].quantile(0.75)
    IQR = Q3 - Q1
    lim_iqr_inf = Q1 - 3 * IQR
    lim_iqr_sup = Q3 + 3 * IQR
    mask_iqr = (data_videos[col] < lim_iqr_inf) | (data_videos[col] > lim_iqr_sup)
    n_iqr = mask_iqr.sum()

    # --- Límites físicos ---
    min_fis, max_fis = limites_fisicos[col]
    mask_fis = (data_videos[col] < min_fis) | (data_videos[col] > max_fis)
    n_fis = mask_fis.sum()

    # --- Solo IQR / Solo físico / Ambos ---
    solo_iqr  = (mask_iqr & ~mask_fis).sum()
    solo_fis  = (mask_fis & ~mask_iqr).sum()
    ambos     = (mask_iqr & mask_fis).sum()

    resultados.append({
        'variable':          col,
        'min_real':          data_videos[col].min(),
        'max_real':          data_videos[col].max(),
        'lim_iqr_inf':       lim_iqr_inf,
        'lim_iqr_sup':       lim_iqr_sup,
        'lim_fis_inf':       min_fis,
        'lim_fis_sup':       max_fis,
        'outliers_iqr':      n_iqr,
        'outliers_iqr_%':    round(n_iqr / len(data_videos) * 100, 3),
        'outliers_fis':      n_fis,
        'outliers_fis_%':    round(n_fis / len(data_videos) * 100, 3),
        'solo_iqr':          solo_iqr,
        'solo_fisico':       solo_fis,
        'ambos':             ambos,
    })

df_comparacion = pd.DataFrame(resultados).set_index('variable')

pd.set_option('display.float_format', '{:.3f}'.format)
print("=== Comparación IQR vs Límites Físicos ===\n")
print(df_comparacion.to_string())
```

**Hallazgos**

Como se comentó, en varias de estar variables, los limites que definió el rango interquartilico fueron cero o incluso valores negativos en el caso de las distancias.

**Tratamiento**

```python
# ============================================================
# TRATAMIENTO DE OUTLIERS - Clipeo por límites físicos
# ============================================================

for col, (min_val, max_val) in limites_fisicos.items():
    if col == 'ball_displacement':
        continue  # sin outliers reales
    data_videos[col] = data_videos[col].clip(lower=min_val, upper=max_val)

# Verificación
print("=== Verificación post-clipeo ===")
for col, (min_val, max_val) in limites_fisicos.items():
    real_min = data_videos[col].min()
    real_max = data_videos[col].max()
    ok = "OK" if real_min >= min_val and real_max <= max_val else "REVISAR"
    print(f"{ok} | {col}: [{real_min:.3f}, {real_max:.3f}]")
```

### Construcción de la variable objetivo y variables de contexto

> **Nota:** El partido con ID 23 finalizó con marcador 7-6 (diferencia de 1 punto),
> lo cual no cumple la regla de diferencia mínima. Se trata como un error de
> registro y no afecta el análisis.

```python
# ============================================================
# CONSTRUCCIÓN DE LA VARIABLE OBJETIVO Y VARIABLES DE CONTEXTO
# ============================================================

data_partidos = data_partidos.sort_values(['ID_PARTIDO', 'ID_PUNTO']).reset_index(drop=True)

# ------------------------------------------------------------
# 1. Marcador acumulado (ya calculado, verificar que existe)
# ------------------------------------------------------------
data_partidos['MARCADOR_EQUIPO_1'] = (
    data_partidos.groupby('ID_PARTIDO')['PUNTO_EQUIPO_1'].cumsum()
)
data_partidos['MARCADOR_EQUIPO_2'] = (
    data_partidos.groupby('ID_PARTIDO')['PUNTO_EQUIPO_2'].cumsum()
)

# ------------------------------------------------------------
# 2. Ganador del partido (variable objetivo)
# ------------------------------------------------------------
ganador_partido = (
    data_partidos.groupby('ID_PARTIDO')
    .last()[['MARCADOR_EQUIPO_1', 'MARCADOR_EQUIPO_2']]
    .apply(lambda row: 1 if row['MARCADOR_EQUIPO_1'] > row['MARCADOR_EQUIPO_2'] else 2, axis=1)
    .rename('EQUIPO_GANADOR')
    .reset_index()
)

data_partidos = data_partidos.merge(ganador_partido, on='ID_PARTIDO')

# ------------------------------------------------------------
# 3. Variables de contexto del marcador
# ------------------------------------------------------------
data_partidos['DIFERENCIA_MARCADOR'] = (
    data_partidos['MARCADOR_EQUIPO_1'] - data_partidos['MARCADOR_EQUIPO_2']
)
data_partidos['PUNTOS_JUGADOS'] = (
    data_partidos['MARCADOR_EQUIPO_1'] + data_partidos['MARCADOR_EQUIPO_2']
)
data_partidos['WINRATE_ACUM_E1'] = (
    data_partidos['MARCADOR_EQUIPO_1'] / data_partidos['PUNTOS_JUGADOS']
)
data_partidos['WINRATE_ACUM_E2'] = (
    data_partidos['MARCADOR_EQUIPO_2'] / data_partidos['PUNTOS_JUGADOS']
)

# ------------------------------------------------------------
# 4. Variables de momentum
# ------------------------------------------------------------
# Racha de puntos consecutivos por equipo
def calcular_racha(serie):
    racha = []
    contador = 0
    for val in serie:
        if val == 1:
            contador += 1
        else:
            contador = 0
        racha.append(contador)
    return racha

data_partidos['RACHA_EQUIPO_1'] = (
    data_partidos.groupby('ID_PARTIDO')['PUNTO_EQUIPO_1']
    .transform(calcular_racha)
)
data_partidos['RACHA_EQUIPO_2'] = (
    data_partidos.groupby('ID_PARTIDO')['PUNTO_EQUIPO_2']
    .transform(calcular_racha)
)

# Puntos ganados en los últimos 5 puntos
data_partidos['PUNTOS_ULTIMOS_5_E1'] = (
    data_partidos.groupby('ID_PARTIDO')['PUNTO_EQUIPO_1']
    .transform(lambda x: x.rolling(5, min_periods=1).sum())
)
data_partidos['PUNTOS_ULTIMOS_5_E2'] = (
    data_partidos.groupby('ID_PARTIDO')['PUNTO_EQUIPO_2']
    .transform(lambda x: x.rolling(5, min_periods=1).sum())
)

# ------------------------------------------------------------
# Verificación
# ------------------------------------------------------------
print(data_partidos[[
    'ID_PARTIDO', 'ID_PUNTO',
    'MARCADOR_EQUIPO_1', 'MARCADOR_EQUIPO_2',
    'DIFERENCIA_MARCADOR', 'PUNTOS_JUGADOS',
    'RACHA_EQUIPO_1', 'RACHA_EQUIPO_2',
    'PUNTOS_ULTIMOS_5_E1', 'PUNTOS_ULTIMOS_5_E2',
    'WINRATE_ACUM_E1', 'WINRATE_ACUM_E2',
    'EQUIPO_GANADOR'
]].head(20).to_string())
```

A partir de `data_partidos` se construyeron las variables necesarias para contextualizar cada punto dentro del partido. Se calculó el marcador acumulado por equipo usando una suma acumulada por partido, lo que permite saber el estado del marcador en el momento exacto en que se jugó cada punto. A partir de este marcador se derivaron variables de contexto como la diferencia de marcador, la proporción de puntos ganados (win rate acumulado) y variables de momentum como la racha de puntos consecutivos y los puntos ganados en los últimos 5 puntos. Adicionalmente se identificó al equipo ganador del partido tomando el estado final del marcador. Estas variables son fundamentales para que el modelo pueda aprender no solo del desempeño físico sino también del contexto competitivo en que se desarrolla cada punto.

> **Nota:** El partido con ID 23 finalizó con marcador 7-6 (diferencia de 1 punto), lo cual no cumple la regla de diferencia mínima de 2. Se trató como un error de registro y no afecta el análisis.


### Validación - Posición del equipo

Ahora que se han agrupado, se quiere validar la posición (top/bottom; left/right) de los miembros del equipó contrastando ambas bases de datos, corrigiendolas dandole prioridad a data_partidos. 

```python
    # ============================================================
    # LOOKUP CORREGIDO: incluye ID_PUNTO para partidos donde
    # la zona cambia entre puntos
    # ============================================================

    lookup = pd.melt(
        data_partidos[[
            'ID_PARTIDO', 'ID_PUNTO',
            'ID_JUGADOR_1_EQUIPO_1', 'ID_JUGADOR_2_EQUIPO_1',
            'ID_JUGADOR_1_EQUIPO_2', 'ID_JUGADOR_2_EQUIPO_2',
            'EQUIPO_1_POSICION', 'EQUIPO_2_POSICION',
            'POSICION_JUGADOR_1_EQUIPO_1', 'POSICION_JUGADOR_2_EQUIPO_1',
            'POSICION_JUGADOR_1_EQUIPO_2', 'POSICION_JUGADOR_2_EQUIPO_2'
        ]],
        id_vars=[
            'ID_PARTIDO', 'ID_PUNTO',
            'EQUIPO_1_POSICION', 'EQUIPO_2_POSICION',
            'POSICION_JUGADOR_1_EQUIPO_1', 'POSICION_JUGADOR_2_EQUIPO_1',
            'POSICION_JUGADOR_1_EQUIPO_2', 'POSICION_JUGADOR_2_EQUIPO_2'
        ],
        value_vars=[
            'ID_JUGADOR_1_EQUIPO_1', 'ID_JUGADOR_2_EQUIPO_1',
            'ID_JUGADOR_1_EQUIPO_2', 'ID_JUGADOR_2_EQUIPO_2'
        ],
        var_name='rol',
        value_name='player_name'
    )

    lookup['team_correcto'] = lookup.apply(
        lambda r: r['EQUIPO_1_POSICION'] if 'EQUIPO_1' in r['rol'] else r['EQUIPO_2_POSICION'],
        axis=1
    ).str.upper()

    lookup['zone_correcta'] = lookup.apply(
        lambda r: (
            r['POSICION_JUGADOR_1_EQUIPO_1'] if r['rol'] == 'ID_JUGADOR_1_EQUIPO_1' else
            r['POSICION_JUGADOR_2_EQUIPO_1'] if r['rol'] == 'ID_JUGADOR_2_EQUIPO_1' else
            r['POSICION_JUGADOR_1_EQUIPO_2'] if r['rol'] == 'ID_JUGADOR_1_EQUIPO_2' else
            r['POSICION_JUGADOR_2_EQUIPO_2']
        ),
        axis=1
    ).str.lower()

    lookup = lookup.rename(columns={
        'ID_PARTIDO': 'partido',
        'ID_PUNTO':   'punto'
    })[['partido', 'punto', 'player_name', 'team_correcto', 'zone_correcta']]

    # Merge por partido + punto + jugador
    data_videos = data_videos.merge(
        lookup, on=['partido', 'punto', 'player_name'], how='left'
    )

    mask_con_match = data_videos['team_correcto'].notna()

    # Diagnóstico de discrepancias ANTES de corregir
    discrepancias_team = (
        mask_con_match &
        (data_videos['team'] != data_videos['team_correcto'])
    ).sum()

    discrepancias_zone = (
        mask_con_match &
        data_videos['zone'].notna() &
        (data_videos['zone'] != data_videos['zone_correcta'])
    ).sum()

    total      = len(data_videos)
    corregidos = mask_con_match.sum()
    sin_match  = (~mask_con_match).sum()

    print(f"Total frames:                  {total:,}")
    print(f"Frames con match en lookup:    {corregidos:,} ({corregidos/total*100:.1f}%)")
    print(f"Sin match (Player_X u otros):  {sin_match:,} ({sin_match/total*100:.1f}%)")
    print(f"\nDiscrepancias en team:         {discrepancias_team:,} ({discrepancias_team/total*100:.2f}%)")
    print(f"Discrepancias en zone:         {discrepancias_zone:,} ({discrepancias_zone/total*100:.2f}%)")

    # Corregir team y zone
    data_videos.loc[mask_con_match, 'team'] = data_videos.loc[mask_con_match, 'team_correcto']
    data_videos.loc[mask_con_match, 'zone'] = data_videos.loc[mask_con_match, 'zone_correcta']

    data_videos = data_videos.drop(columns=['team_correcto', 'zone_correcta'])

    print(f"\nShape final: {data_videos.shape}")
```

### Renombrar variables de equipo (Perspectiva dual)

```python
# ============================================================
# BLOQUE 1: RENOMBRAR VARIABLES DE EQUIPO → PERSPECTIVA DUAL
# ============================================================
#
# Estrategia: generar DOS filas por punto.
#   - Fila A: perspectiva del EQUIPO_1  → equipo=E1, rival=E2
#   - Fila B: perspectiva del EQUIPO_2  → equipo=E2, rival=E1
#
# Esto elimina el sesgo de posición (BOTTOM/TOP) y permite que
# cualquier modelo aprenda patrones desde la perspectiva de
# "mi equipo" sin importar qué bando ocupó en cancha.
# ============================================================

def construir_perspectiva(data, equipo_focal, equipo_rival):
    """
    Crea un subconjunto del dataframe desde la perspectiva de `equipo_focal`.
    equipo_focal / equipo_rival: 1 ó 2
    """
    sufijo_f = f"_e{equipo_focal}"   # columnas del equipo focal
    sufijo_r = f"_e{equipo_rival}"   # columnas del rival

    df = data.copy()

    # ── Jugadores ──────────────────────────────────────────
    df['jugador_1_equipo']  = df[f'ID_JUGADOR_1_EQUIPO_{equipo_focal}']
    df['jugador_2_equipo']  = df[f'ID_JUGADOR_2_EQUIPO_{equipo_focal}']
    df['jugador_1_rival']   = df[f'ID_JUGADOR_1_EQUIPO_{equipo_rival}']
    df['jugador_2_rival']   = df[f'ID_JUGADOR_2_EQUIPO_{equipo_rival}']

    # ── Estadísticas del partido (por punto) ───────────────
    df['velocidad_prom_equipo']      = df[f'velocidad_prom{sufijo_f}']
    df['velocidad_prom_rival']       = df[f'velocidad_prom{sufijo_r}']
    df['velocidad_max_equipo']       = df[f'velocidad_max{sufijo_f}']
    df['velocidad_max_rival']        = df[f'velocidad_max{sufijo_r}']
    df['desplazamiento_equipo']      = df[f'desplazamiento_total{sufijo_f}']
    df['desplazamiento_rival']       = df[f'desplazamiento_total{sufijo_r}']
    df['dist_red_prom_equipo']       = df[f'dist_red_prom{sufijo_f}']
    df['dist_red_prom_rival']        = df[f'dist_red_prom{sufijo_r}']
    df['dist_companero_prom_equipo'] = df[f'dist_companero_prom{sufijo_f}']
    df['dist_companero_prom_rival']  = df[f'dist_companero_prom{sufijo_r}']
    df['hits_equipo']                = df[f'hits{sufijo_f}']
    df['hits_rival']                 = df[f'hits{sufijo_r}']
    df['aceleracion_prom_equipo']    = df[f'aceleracion_prom{sufijo_f}']
    df['aceleracion_prom_rival']     = df[f'aceleracion_prom{sufijo_r}']

    # ── Marcador desde perspectiva del equipo focal ─────────
    df['marcador_equipo']  = df[f'MARCADOR_EQUIPO_{equipo_focal}']
    df['marcador_rival']   = df[f'MARCADOR_EQUIPO_{equipo_rival}']
    df['gano_punto']       = (df['ganador_punto'] == equipo_focal).astype(int)  # 1 si ganó, 0 si no

    # ── Cancha (posición en la cancha este partido) ─────────
    df['cancha'] = df['CANCHA']

    # ── Columna auxiliar para identificar la perspectiva ────
    df['perspectiva_equipo'] = equipo_focal  # 1 ó 2; útil para trazabilidad

    # Seleccionar solo las columnas relevantes
    cols_comunes = [
        'partido', 'punto', 'cancha', 'perspectiva_equipo',
        'jugador_1_equipo', 'jugador_2_equipo',
        'jugador_1_rival',  'jugador_2_rival',
        'duracion_punto',
        # estadísticas del punto
        'velocidad_prom_equipo',      'velocidad_prom_rival',
        'velocidad_max_equipo',       'velocidad_max_rival',
        'desplazamiento_equipo',      'desplazamiento_rival',
        'dist_red_prom_equipo',       'dist_red_prom_rival',
        'dist_companero_prom_equipo', 'dist_companero_prom_rival',
        'hits_equipo',                'hits_rival',
        'aceleracion_prom_equipo',    'aceleracion_prom_rival',
        'velocidad_prom_pelota',      'velocidad_max_pelota',
        # marcador
        'marcador_equipo', 'marcador_rival',
        'gano_punto',
    ]
    return df[cols_comunes]


# Construir las dos perspectivas y apilarlas
df_e1 = construir_perspectiva(data, equipo_focal=1, equipo_rival=2)
df_e2 = construir_perspectiva(data, equipo_focal=2, equipo_rival=1)

data_dual = (
    pd.concat([df_e1, df_e2], ignore_index=True)
    .sort_values(['partido', 'punto', 'perspectiva_equipo'])
    .reset_index(drop=True)
)

print(f"Shape data_dual (perspectiva doble): {data_dual.shape}")
print(data_dual.head(6).to_string())
```

Para eliminar el sesgo de posición (qué equipo ocupó TOP o BOTTOM en la cancha) y permitir que el modelo aprenda patrones desde la perspectiva de "mi equipo", se duplicó cada punto generando dos filas: una desde la perspectiva del equipo 1 y otra desde la del equipo 2. En cada fila las variables se renombraron como `_equipo` (equipo focal) y `_rival` (equipo contrario), y la variable objetivo `gano_punto` toma el valor 1 si el equipo focal ganó ese punto. Esta estrategia también duplica el tamaño del dataset de entrenamiento, lo cual es beneficioso dado el volumen limitado de partidos disponibles.

### Variables acumuladas durante el partido

```python
# ============================================================
# BLOQUE 2: VARIABLES ACUMULADAS DURANTE EL PARTIDO
# ============================================================
#
# Para cada fila (partido × punto × perspectiva_equipo),
# calculamos estadísticas acumuladas *hasta ese punto*
# (excluyendo el punto actual → expanding con shift(1)).
#
# Esto evita fuga de información: al inicio del punto N solo
# conocemos lo ocurrido en los puntos 1…N-1.
# ============================================================

# Ordenar correctamente antes de calcular acumulados
data_dual = data_dual.sort_values(['partido', 'perspectiva_equipo', 'punto']).reset_index(drop=True)

grupo = data_dual.groupby(['partido', 'perspectiva_equipo'])


def acum_expandido(series):
    """Promedio acumulado excluyendo el punto actual (shift 1)."""
    return series.expanding().mean().shift(1)

def acum_sum_expandido(series):
    """Suma acumulada excluyendo el punto actual."""
    return series.expanding().sum().shift(1)

def acum_max_expandido(series):
    """Máximo acumulado excluyendo el punto actual."""
    return series.expanding().max().shift(1)


# ── 2.1 Velocidad promedio acumulada ───────────────────────
data_dual['vel_prom_acum_equipo'] = grupo['velocidad_prom_equipo'].transform(acum_expandido)
data_dual['vel_prom_acum_rival']  = grupo['velocidad_prom_rival'].transform(acum_expandido)

# ── 2.2 Velocidad máxima acumulada (pico histórico del partido) ─
data_dual['vel_max_acum_equipo'] = grupo['velocidad_max_equipo'].transform(acum_max_expandido)
data_dual['vel_max_acum_rival']  = grupo['velocidad_max_rival'].transform(acum_max_expandido)

# ── 2.3 Desplazamiento total acumulado ─────────────────────
data_dual['desplaz_acum_equipo'] = grupo['desplazamiento_equipo'].transform(acum_sum_expandido)
data_dual['desplaz_acum_rival']  = grupo['desplazamiento_rival'].transform(acum_sum_expandido)

# ── 2.4 Golpes acumulados ───────────────────────────────────
data_dual['hits_acum_equipo'] = grupo['hits_equipo'].transform(acum_sum_expandido)
data_dual['hits_acum_rival']  = grupo['hits_rival'].transform(acum_sum_expandido)

# ── 2.5 Aceleración promedio acumulada ─────────────────────
data_dual['acel_prom_acum_equipo'] = grupo['aceleracion_prom_equipo'].transform(acum_expandido)
data_dual['acel_prom_acum_rival']  = grupo['aceleracion_prom_rival'].transform(acum_expandido)

# ── 2.6 Distancia a la red acumulada ───────────────────────
data_dual['dist_red_acum_equipo'] = grupo['dist_red_prom_equipo'].transform(acum_expandido)
data_dual['dist_red_acum_rival']  = grupo['dist_red_prom_rival'].transform(acum_expandido)

# ── 2.7 Distancia al compañero acumulada ───────────────────
data_dual['dist_comp_acum_equipo'] = grupo['dist_companero_prom_equipo'].transform(acum_expandido)
data_dual['dist_comp_acum_rival']  = grupo['dist_companero_prom_rival'].transform(acum_expandido)

# ── 2.8 Win rate acumulado en el partido ───────────────────
#    (% de puntos ganados hasta ahora en este partido)
data_dual['win_rate_acum_equipo'] = grupo['gano_punto'].transform(acum_expandido)

# ── 2.9 Racha reciente: puntos ganados en los últimos 3 ────
def racha_n(series, n=3):
    return series.shift(1).rolling(window=n, min_periods=1).sum()

data_dual['racha_ultimos3_equipo'] = grupo['gano_punto'].transform(lambda x: racha_n(x, 3))
data_dual['racha_ultimos3_rival']  = grupo['gano_punto'].transform(
    lambda x: racha_n(1 - x, 3)   # puntos ganados por el rival = perdidos por focal
)

# ── 2.10 Diferencia de marcador (perspectiva equipo focal) ─
#    Positivo = equipo va ganando, negativo = va perdiendo
data_dual['diferencia_marcador'] = data_dual['marcador_equipo'] - data_dual['marcador_rival']

# ── 2.11 Velocidad de la pelota acumulada (promedio) ───────
data_dual['vel_pelota_acum'] = grupo['velocidad_prom_pelota'].transform(acum_expandido)


# ============================================================
# VERIFICACIÓN RÁPIDA
# ============================================================

print("\n=== Shape final ===")
print(data_dual.shape)

print("\n=== Columnas ===")
print(data_dual.columns.tolist())

print("\n=== Primeros 4 puntos del partido 1, perspectiva equipo 1 ===")
muestra = data_dual[
    (data_dual['partido'] == 1) & (data_dual['perspectiva_equipo'] == 1)
].head(4)
print(muestra[['partido', 'punto', 'velocidad_prom_equipo',
               'vel_prom_acum_equipo', 'hits_acum_equipo',
               'racha_ultimos3_equipo', 'diferencia_marcador',
               'gano_punto']].to_string())

print("\n=== Nulos por columna (acumuladas) ===")
cols_acum = [c for c in data_dual.columns if 'acum' in c or 'racha' in c or 'diferencia' in c]
print(data_dual[cols_acum].isnull().sum())
```

Se construyeron variables acumuladas que capturan el historial del partido hasta el punto anterior, evitando fuga de información (data leakage). Para cada punto N, las variables acumuladas reflejan únicamente lo ocurrido en los puntos 1 a N-1, implementado mediante `expanding().mean().shift(1)`. Se calcularon promedios acumulados de velocidad, aceleración y distancias, sumas acumuladas de golpes y desplazamiento, el win rate acumulado del equipo focal, y una racha reciente de los últimos 3 puntos. Estas variables son clave para que el modelo capture el momentum y la evolución del rendimiento a lo largo del partido.

#### Imputación nulos primer punto

```python
# ============================================================
# IMPUTACIÓN DE NULOS DEL PRIMER PUNTO (60 filas)
# Estrategia:
#   - Variables de conteo/proporción → 0 (sin historial previo)
#   - Variables de velocidad/distancia → mediana del partido
#   - Indicador es_primer_punto para que el modelo lo sepa
# ============================================================

# ------------------------------------------------------------
# 1. Crear variable indicadora de primer punto
# ------------------------------------------------------------
data_dual['es_primer_punto'] = (
    data_dual
    .groupby(['partido', 'perspectiva_equipo'])['punto']
    .transform(lambda x: (x == x.min()).astype(int))
)

print(f"Filas marcadas como primer punto: {data_dual['es_primer_punto'].sum()}")

# ------------------------------------------------------------
# 2. Definir qué columnas van con 0 y cuáles con mediana
# ------------------------------------------------------------

# Con 0: semántica clara → antes del primer punto no hay
# golpes, rachas, puntos ganados ni win rate acumulado
cols_imputar_cero = [
    'hits_acum_equipo',
    'hits_acum_rival',
    'win_rate_acum_equipo',
    'racha_ultimos3_equipo',
    'racha_ultimos3_rival',
    'desplaz_acum_equipo',
    'desplaz_acum_rival',
    'vel_pelota_acum',
]

# Con mediana del partido: variables físicas donde 0 no tiene
# sentido (los jugadores sí se mueven en el primer punto)
cols_imputar_mediana = [
    'vel_prom_acum_equipo',
    'vel_prom_acum_rival',
    'vel_max_acum_equipo',
    'vel_max_acum_rival',
    'acel_prom_acum_equipo',
    'acel_prom_acum_rival',
    'dist_red_acum_equipo',
    'dist_red_acum_rival',
    'dist_comp_acum_equipo',
    'dist_comp_acum_rival',
]

# ------------------------------------------------------------
# 3. Imputar con 0
# ------------------------------------------------------------
for col in cols_imputar_cero:
    nulos_antes = data_dual[col].isna().sum()
    data_dual[col] = data_dual[col].fillna(0)
    print(f"  {col}: {nulos_antes} nulos → imputados con 0")

# ------------------------------------------------------------
# 4. Imputar con mediana del partido (por perspectiva_equipo)
#    Se calcula la mediana excluyendo las filas nulas (primer punto)
#    usando solo los puntos que sí tienen historial
# ------------------------------------------------------------
for col in cols_imputar_mediana:
    nulos_antes = data_dual[col].isna().sum()

    mediana_partido = (
        data_dual
        .groupby(['partido', 'perspectiva_equipo'])[col]
        .transform(lambda x: x.fillna(x.median()))
    )
    data_dual[col] = data_dual[col].fillna(mediana_partido)

    # Si el partido solo tiene 1 punto (mediana también es NaN),
    # caer a mediana global de la columna
    mediana_global = data_dual[col].median()
    data_dual[col] = data_dual[col].fillna(mediana_global)

    print(f"  {col}: {nulos_antes} nulos → imputados con mediana del partido")

# ------------------------------------------------------------
# 5. Verificación final
# ------------------------------------------------------------
cols_acum = cols_imputar_cero + cols_imputar_mediana
nulos_post = data_dual[cols_acum].isnull().sum()

print(f"\n=== Verificación post-imputación ===")
print(f"Nulos restantes en columnas acumuladas:")
print(nulos_post[nulos_post > 0] if nulos_post.sum() > 0 else "  Sin nulos restantes ✓")

print(f"\nNulos totales en data_dual:")
nulos_totales = data_dual.isnull().sum()
print(nulos_totales[nulos_totales > 0] if nulos_totales.sum() > 0 else "  Sin nulos en todo el dataframe ✓")

print(f"\nShape final: {data_dual.shape}")
print(f"\nDistribución es_primer_punto:\n{data_dual['es_primer_punto'].value_counts()}")
```

Las variables acumuladas son estructuralmente nulas para el primer punto de cada partido, ya que no existe historial previo. Se aplicó una estrategia de imputación diferenciada: las variables de conteo y proporción (golpes, win rate, racha, desplazamiento) se imputaron con 0, reflejando que no hay historial previo; las variables físicas donde 0 no tiene sentido semántico (velocidades, distancias) se imputaron con la mediana del partido. Adicionalmente se creó la variable indicadora `es_primer_punto` para que el modelo pueda distinguir estos casos durante el entrenamiento.

Se construyeron dos versiones del dataset final con el objetivo de evaluar si la información de perfil de los jugadores como deportistas aporta valor predictivo al modelo, o si las métricas físicas capturadas por el tracking de video son suficientes por sí solas.data_sin_jugadores contiene únicamente las variables derivadas del video y del marcador, sin ninguna información sobre los jugadores. Este dataframe permite entrenar y evaluar modelos basados exclusivamente en el desempeño observable durante el partido.data_con_jugadores incorpora adicionalmente el perfil deportivo de cada uno de los cuatro jugadores involucrados en el punto: nivel de pádel, experiencia, condición física, frecuencia de entrenamiento, entre otras. Dado que 9 de los 36 jugadores no diligenciaron el formulario de perfil, sus valores se imputaron con la mediana (variables numéricas) o la moda (variables categóricas) calculadas sobre los jugadores que sí tienen perfil.Comparar el desempeño de modelos entrenados con cada versión permitirá determinar si características intrínsecas de los jugadores como su nivel declarado o su condición física tienen poder explicativo adicional más allá de lo que ya captura el tracking, o si por el contrario el comportamiento en cancha es suficiente para predecir el resultado del punto.
